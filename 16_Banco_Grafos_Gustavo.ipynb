### 2: Modelagem bsica
- **Ns**: `:Pessoa`, `:Produto`, `:Node` (genrico)  
- **Relacionamentos**: `:AMIGO_DE`, `:COMPROU`, `:CONNECTED`  
- Use propriedades em ns/relacionamentos (`nome`, `idade`, `peso`, `timestamp`).  
- Crie ndices/constraints em propriedades frequentemente buscadas (ex: `:Pessoa(nome)`, `:Product(sku)`).

````python
driver = GraphDatabase.driver(NEO4J_URI, auth=basic_auth(NEO4J_USERNAME, NEO4J_PASSWORD))
````

````python
with driver.session() as s:  # Abre uma sesso com o banco Neo4j e a referencia como 's'
    
    # Apaga todos os ns do grafo e seus relacionamentos
    s.run("MATCH (n) DETACH DELETE n") 
    
    # Cria uma regra de restrio de unicidade para o campo 'nome' em ns do tipo Pessoa, se ainda no existir
    s.run("""CREATE CONSTRAINT pessoa_nome_unico IF NOT EXISTS
        FOR (p:Pessoa)
        REQUIRE p.nome IS UNIQUE
    """) 

    # Cria um ndice para o campo 'id' em ns do tipo Node, se ainda no existir
    s.run("""
        CREATE INDEX node_id_index IF NOT EXISTS
        FOR (n:Node)
        ON (n.id)
    """) 

print("Limpeza de ns e ndices criados.")  # Informa no console que a limpeza e a criao de ndices foram concludas
````

#### Constraints

So **regras de integridade** aplicadas ao grafo.
Elas garantem que os dados obedeam certas condies.

Em resumo:
- servem para garantir consistncia
- impedem duplicaes, propriedades faltantes ou violaes de regras
- algumas criam ndices automaticamente

````python
with driver.session() as s:
    constraints = s.run("SHOW CONSTRAINTS").data()

df_constraints = pd.DataFrame(constraints)
df_constraints
````

#### Indexes

So **estruturas de acelerao de busca**.
Eles permitem que o Neo4j encontre ns ou relacionamentos rapidamente, sem percorrer todo o grafo.

Em resumo:

- servem para otimizar performance
- aceleram MATCH, WHERE e MERGE
- funcionam como o ndice de um livro

````python
with driver.session() as s:
    indexes = s.run("SHOW INDEXES").data()

df_indexes = pd.DataFrame(indexes)
df_indexes
````

### E se o nmero de ns for muito grande?

````python
def criar_grafo_sintetico(driver, n_pessoas=100, seed=42):
    random.seed(seed)

    profissoes = [
        "Engenheiro de Dados",
        "Analista de Sistemas",
        "Cientista de Dados",
        "Gestor de Projetos",
        "Analista de Qualidade",
        "Tcnico de Manuteno",
        "Professor",
        "Consultor",
    ]

    cidades = [
        "So Paulo",
        "Campinas",
        "Jundia",
        "Santos",
        "So Jos dos Campos",
        "Sorocaba",
    ]

    interesses = [
        "Machine Learning",
        "IoT",
        "Big Data",
        "Viso Computacional",
        "DevOps",
        "Automao Industrial",
        "Gesto gil",
        "Analytics",
    ]

    tipos_rel = ["AMIGO_DE", "COLEGA", "FAMILIAR", "CONHECE"]

    with driver.session() as s:
        # Limpar banco (cuidado, aqui apaga tudo)
        s.run("MATCH (n) DETACH DELETE n")

        # Constraint de unicidade em nome, padro Neo4j 5
        s.run("""
            CREATE CONSTRAINT pessoa_nome_unico IF NOT EXISTS
            FOR (p:Pessoa)
            REQUIRE p.nome IS UNIQUE
        """)

        # Gerar lista de pessoas com propriedades
        pessoas = []
        for i in range(1, n_pessoas + 1):
            nome = f"Pessoa{i}"
            idade = random.randint(20, 60)
            profissao = random.choice(profissoes)
            cidade = random.choice(cidades)
            grupo = random.randint(1, 4)  # comunidades 1 a 4
            inter = random.sample(interesses, k=random.randint(1, 3))
            pessoas.append(
                {
                    "nome": nome,
                    "idade": idade,
                    "profissao": profissao,
                    "cidade": cidade,
                    "grupo": grupo,
                    "interesses": inter,
                }
            )

        # Criar ns no Neo4j
        for p in pessoas:
            s.run(
                """
                CREATE (:Pessoa {
                    nome: $nome,
                    idade: $idade,
                    profissao: $profissao,
                    cidade: $cidade,
                    grupo: $grupo,
                    interesses: $interesses
                })
                """,
                **p
            )

        # Criar relacionamentos
        # Mais chance de ligao dentro do mesmo grupo
        rel_count = 0
        base_date = date(2020, 1, 1)

        for i in range(len(pessoas)):
            for j in range(i + 1, len(pessoas)):
                p1 = pessoas[i]
                p2 = pessoas[j]

                mesma_comunidade = p1["grupo"] == p2["grupo"]
                if mesma_comunidade:
                    prob = 0.18  # probabilidade maior dentro do grupo
                else:
                    prob = 0.04  # probabilidade menor entre grupos

                if random.random() < prob:
                    tipo = random.choice(tipos_rel)
                    dias = random.randint(0, 365 * 5)
                    d = base_date + timedelta(days=dias)

                    query = f"""
                    MATCH (a:Pessoa {{nome: $nome1}}),
                          (b:Pessoa {{nome: $nome2}})
                    CREATE (a)-[:{tipo} {{desde: date($d)}}]->(b)
                    """

                    s.run(
                        query,
                        nome1=p1["nome"],
                        nome2=p2["nome"],
                        d=d.isoformat()
                    )
                    rel_count += 1

        print(f"Grafo sinttico criado com {len(pessoas)} ns e {rel_count} relacionamentos.")

# Chamar a funo
criar_grafo_sintetico(driver, n_pessoas=100, seed=42)
````

## E se o nmero de ns for realmente grande?
### Estratgia de insero em massa

Para inserir um grafo sinttico grande (ex.: 10k ns, 50k arestas), h duas abordagens:

1. **Via Python + bolt** (usando transaes em lotes):
   - Gera ns em batches (ex.: 1000 por transao).
   - Para arestas, usar matching por propriedade indexada (ex.: `id`) ou criar mapas temporrios.

2. **Import em massa** (`neo4j-admin import`):
   - Gera CSVs de ns/arestas e usa `neo4j-admin import` (muito mais rpido para datasets grandes).
   
Aqui, vamos fazer uma implementao via Python com `networkx` para gerar o grafo e inserir em batches.

````python
from tqdm import tqdm                                  # Importa tqdm, biblioteca usada para criar barras de progresso em loops

# Parmetros (ajuste conforme memria/tempo)
N_NODES = 10000                                        # Define a quantidade de ns que o grafo sinttico ter
N_EDGES = 50000                                        # Define a quantidade total de arestas a serem criadas no grafo

# Cria um grafo aleatrio usando o modelo ErdsRnyi (G(n,m))
G = nx.gnm_random_graph(n=N_NODES,                     # n = nmero de ns
                        m=N_EDGES,                     # m = nmero de arestas
                        seed=42,                       # seed = semente para reprodutibilidade (gera o mesmo grafo sempre)
                        directed=False)                # directed=False significa que o grafo  no-direcionado
                                                     
                                                     
# Exibe quantos ns e quantas arestas foram realmente gerados pelo grafo                    
print("Ns:", G.number_of_nodes(), "Arestas:", G.number_of_edges())                             
````

### Inserir ns em batches usando UNWIND (mais eficiente que criar um por um)

````python
BATCH_SIZE = 1000  # Define o tamanho do lote (batch) de ns que ser enviado ao Neo4j por operao

with driver.session() as s:                           
    nodes = [{"id": int(n)} for n in range(G.number_of_nodes())] # Cria uma lista de dicionrios, cada um representando um n com campo 'id'
                                                                                          
    # Usa tqdm para exibir uma barra de progresso enquanto itera sobre os lotes
    # range percorre de 0 ao total de ns, pulando de BATCH_SIZE em BATCH_SIZE
    for i in tqdm(range(0, len(nodes), BATCH_SIZE), desc="Inserindo ns (batches)"):
        batch = nodes[i:i+BATCH_SIZE]  # Seleciona um lote (slice) da lista de ns. Exemplo: ns 0999, depois 10001999, e assim por diante
        
        # UNWIND Expande a lista 'batch' em linhas individuais.
        s.run("UNWIND $batch AS row CREATE (n:Node {id: row.id})", batch=batch) # Envia o parmetro 'batch' para dentro da query Cypher                            
 
print("Insero de ns concluda.") 
````

### Inserir arestas usando UNWIND + MATCH

````python
edges = list(G.edges())  # Converte todas as arestas do grafo G em uma lista de tuplas (u, v). Cada tupla representa uma aresta entre os ns u e v
BATCH_SIZE = 5000  # Define o nmero de arestas que sero enviadas ao Neo4j por operao (lote)
                                                     
with driver.session() as s:                          

    for i in tqdm(range(0, len(edges), BATCH_SIZE), desc="Inserindo arestas (batches)"):
        batch = [{"u": int(u), "v": int(v)} for u,v in edges[i:i+BATCH_SIZE]]  #Cria um lote de dicionrios, onde cada dicionrio representa uma aresta
        s.run("UNWIND $batch AS row MATCH (a:Node {id: row.u}), (b:Node {id: row.v}) CREATE (a)-[:CONNECTED]->(b)", batch=batch)    
print("Insero de arestas concluda.")
````

### Reconstruo do grafo a partir da base Neo4j

````python
G_db = nx.Graph()                        # Cria um grafo no direcionado em memria, usando NetworkX,
with driver.session() as s:              
    # Define a consulta Cypher que encontra todas as arestas CONNECTED entre ns Node e retorna os ids dos ns origem (source) e destino (target).
    query = """
    MATCH (a:Node)-[:CONNECTED]->(b:Node)
    RETURN a.id AS source, b.id AS target
    """                                  
    result = s.run(query)                # Executa a query no Neo4j e retorna um iterador de registros.

    for record in result:                # Percorre cada registro retornado pela consulta.
        src = record["source"]           # Extrai o valor da coluna 'source' (id do n origem).
        tgt = record["target"]           # Extrai o valor da coluna 'target' (id do n destino).
        G_db.add_edge(src, tgt)          # Adiciona uma aresta no grafo NetworkX entre os ns src e tgt,
                                         # recriando no Python a mesma estrutura presente no Neo4j.

print("Ns em G_db:", G_db.number_of_nodes())        # Exibe quantos ns o grafo reconstrudo possui.
print("Arestas em G_db:", G_db.number_of_edges())    # Exibe quantas arestas o grafo reconstrudo possui.
````

### Anlise do grafo reconstrudo (G_db)

````python
import numpy as np

degrees = [deg for _, deg in G_db.degree()] # Percorre todos os ns do grafo G_db e coleta apenas os graus (o nmero de conexes de cada n).
print("Nmero de ns:", G_db.number_of_nodes())
print("Nmero de arestas:", G_db.number_of_edges())
print("Grau mdio:", np.mean(degrees)) # Calcula a mdia dos graus (conectividade mdia dos ns)
print("Grau mximo:", np.max(degrees)) # Calcula o maior grau encontrado entre todos os ns
print("Grau mnimo:", np.min(degrees)) # Calcula o menor grau encontrado entre todos os ns
````

### Exerccio: Recomendao simples (friends-of-friends)
Enunciado: para um n `X`, retorne ns que so amigos de amigos (2 saltos) que no sejam j vizinhos diretos.


````python
def friends_of_friends(node_id, limit=20):
    with driver.session() as s:
        res = s.run("""
            MATCH (a:Node {id:$id})-[:CONNECTED]->()-[:CONNECTED]->(fof)
            WHERE NOT (a)-[:CONNECTED]->(fof) AND fof.id <> $id
            RETURN DISTINCT fof.id AS fof_id LIMIT $limit
        """, id=node_id, limit=limit)
        return [r["fof_id"] for r in res]

print("Friends of friends para 0:", friends_of_friends(0, limit=10))
````

### Visualizao
- Destacar o n alvo em uma cor.
- Destacar vizinhos diretos em outra cor.
- Destacar friends of friends recomendados em uma terceira cor.
- Visualizar apenas o subgrafo local para no virar uma bola de pelos

````python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_friends_of_friends(node_id, limit_fof=20):
    with driver.session() as s:
        # 1. Buscar n alvo, vizinhos diretos e friends of friends
        data = s.run("""
            MATCH (a:Node {id:$id})
            OPTIONAL MATCH (a)-[:CONNECTED]->(n:Node)
            OPTIONAL MATCH (a)-[:CONNECTED]->(:Node)-[:CONNECTED]->(fof:Node)
            WHERE fof.id <> $id AND NOT (a)-[:CONNECTED]->(fof)
            RETURN a.id AS a_id,
                   collect(DISTINCT n.id) AS neighbors,
                   collect(DISTINCT fof.id) AS fofs
        """, id=node_id).single()

        if not data:
            print(f"N {node_id} no encontrado.")
            return

        a_id = data["a_id"]
        neighbors = data["neighbors"] or []
        fofs = (data["fofs"] or [])[:limit_fof]

        # Conjunto de ns que vamos mostrar
        all_nodes = set([a_id]) | set(neighbors) | set(fofs)

        # 2. Buscar arestas apenas entre esses ns para montar o subgrafo
        rels = s.run("""
            MATCH (x:Node)-[:CONNECTED]->(y:Node)
            WHERE x.id IN $ids AND y.id IN $ids
            RETURN x.id AS source, y.id AS target
        """, ids=list(all_nodes)).data()

    # 3. Construir grafo em memria
    G = nx.Graph()
    for nid in all_nodes:
        if nid == a_id:
            role = "target"
        elif nid in neighbors:
            role = "neighbor"
        elif nid in fofs:
            role = "fof"
        else:
            role = "other"

        G.add_node(nid, role=role)

    for r in rels:
        G.add_edge(r["source"], r["target"])

    # 4. Definir cores por tipo de n
    color_map = []
    for n in G.nodes(data=True):
        role = n[1].get("role")
        if role == "target":
            color_map.append("red")        # n alvo
        elif role == "neighbor":
            color_map.append("lightblue")  # vizinho direto
        elif role == "fof":
            color_map.append("lightgreen") # recomendao
        else:
            color_map.append("gray")

    # 5. Layout e plotagem
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G, k=0.3, iterations=50)

    nx.draw(
        G,
        pos,
        node_color=color_map,
        node_size=120,
        width=0.6,
        with_labels=True,
        font_size=7
    )

    # Pequena legenda manual
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor="red",        label="N alvo"),
        Patch(facecolor="lightblue",  label="Vizinhos diretos"),
        Patch(facecolor="lightgreen", label="Friends of friends (recomendados)"),
    ]
    plt.legend(handles=legend_elements, loc="best")

    plt.title(f"Friends of friends para o n {node_id}")
    plt.axis("off")
    plt.show()
````

````python
print("Friends of friends para 0:", friends_of_friends(0, limit=10))
visualize_friends_of_friends(0, limit_fof=10)
````
